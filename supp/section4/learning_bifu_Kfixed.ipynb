```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
import shelve
import string
import numba as nb
import timeit
import matplotlib.cm as cm
from matplotlib.colors import Normalize
from matplotlib import colors
```

```python
Flist = []
Rlist = []
Plist = []
X0list = []
Y0list = []
Clist = []
Slist =[]
with shelve.open(f'shelve/Fbifu11.shelve', 'r') as shelf:
    for F in shelf.keys():

        Plist.append(shelf[F][3])
        Flist.append(shelf[F][0])
        Rlist.append(shelf[F][2])
        X0list.append(shelf[F][4])
        Y0list.append(shelf[F][5])
        Clist.append(shelf[F][7])
        Slist.append(shelf[F][8])

m=2
Farray= np.array(Flist[::m]) 
Rarray= np.array(Rlist[::m]) 
Parray= np.array(Plist[::m]) 
Carray= np.array(Clist[::m])
Sarray= np.floor(np.abs(np.array(Slist[::m])))
```

```python
Rmin  =  np.array([  min(np.sqrt(x0**2 + y0**2)[-10_000:])  for  x0,y0 in zip(X0list[::m],Y0list[::m])   ])
Rmax  =  np.array([  max(np.sqrt(x0**2 + y0**2)[-10_000:])  for  x0,y0 in zip(X0list[::m],Y0list[::m])   ])
X0_ct = np.array([  np.round(np.linalg.norm(x0[-1500:-1]-x0[-1499:],2),2)==0  for  x0 in X0list[::m]   ])
Y0_ct = np.array([  np.round(np.linalg.norm(y0[-1500:-1]-y0[-1499:],2),2)==0  for  y0 in Y0list[::m]   ])
Cb = Carray>0
Sb = Sarray>0
```

```python

fig = plt.figure(figsize=(7,3))
frame1 = plt.gca()
plt.scatter(Farray[(Y0_ct + X0_ct)],Rmin[(Y0_ct + X0_ct)], color='black',marker='^', s=45, alpha=0.5)
plt.scatter(Farray[(Y0_ct + X0_ct)],Rmax[(Y0_ct + X0_ct)], color='black',marker='v', s=45, alpha=0.5)

plt.scatter(Farray[  Parray  & ((Y0_ct + X0_ct)==False)& (Sb == False)],Rmin[  Parray  & ((Y0_ct + X0_ct)==False) & (Sb == False)], color='purple',marker='^', s=45, alpha=0.5)
plt.scatter(Farray[  Parray  & ((Y0_ct + X0_ct)==False) & (Sb == False)],Rmax[  Parray  & ((Y0_ct + X0_ct)==False)& (Sb == False) ], color='purple',marker='v', s=45, alpha=0.5)

plt.scatter(Farray[  Parray  & ((Y0_ct + X0_ct)==False)& (Sb)],Rmin[  Parray  & ((Y0_ct + X0_ct)==False) & (Sb)], color='blue',marker='^', s=45, alpha=0.5)
plt.scatter(Farray[  Parray  & ((Y0_ct + X0_ct)==False) & (Sb)],Rmax[  Parray  & ((Y0_ct + X0_ct)==False)& (Sb) ], color='blue',marker='v', s=45, alpha=0.5)




plt.scatter(Farray[ (Parray == False)  & ((Y0_ct + X0_ct)==False) & (Sb == False)    ],\
             Rmax[ (Parray == False)  & ((Y0_ct + X0_ct)==False)   & (Sb == False ) ], color='red',marker='v', s=45, alpha=0.5)

plt.scatter(Farray[ (Parray == False)  & ((Y0_ct + X0_ct)==False) & (Sb == False)    ],\
             Rmin[ (Parray == False)  & ((Y0_ct + X0_ct)==False)   & (Sb == False ) ], color='red',marker='^', s=45, alpha=0.5)

plt.scatter(Farray[Sb & (Parray == False) ],Rmin[Sb & (Parray == False)], color='cyan',marker='^', s=45, alpha=0.5)
plt.scatter(Farray[Sb & (Parray == False) ],Rmax[Sb & (Parray == False)], color='cyan',marker='v', s=45, alpha=0.5)

#plt.axis('off')
#frame1.axes.get_xaxis().set_visible(False)
#frame1.axes.get_yaxis().set_visible(False)
#plt.xlabel('F')
#plt.ylabel('r')
plt.ylim([0,1.05])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

```
